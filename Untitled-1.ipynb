```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from lightgbm import LGBMRegressor
import matplotlib.pyplot as plt
import joblib

# Load and preprocess data
data = pd.read_excel("ديزل 20-24.xlsx")
print("Initial data shape:", data.shape)

# Handle missing values
data = data.dropna(subset=["المنطقة", "رقم الحساب", "اسم الزبون", "اسم المنتج", "وصف التعبئة", "السنة", "الشهر", "الكمية باللتر"])
print("Shape after dropping missing values:", data.shape)

# Ensure data types
data["المنطقة"] = data["المنطقة"].astype(str)
data["رقم الحساب"] = data["رقم الحساب"].astype(str)
data["اسم الزبون"] = data["اسم الزبون"].astype(str)
data["اسم المنتج"] = data["اسم المنتج"].astype(str)
data["وصف التعبئة"] = data["وصف التعبئة"].astype(str)
data["السنة"] = data["السنة"].astype(int)
data["الشهر"] = data["الشهر"].astype(int)
data["الكمية باللتر"] = data["الكمية باللتر"].clip(lower=0).astype(float)

# Add seasonality features
data["month_sin"] = np.sin(2 * np.pi * data["الشهر"] / 12)
data["month_cos"] = np.cos(2 * np.pi * data["الشهر"] / 12)

# Encode categorical variables
label_encoders = {}
categorical_columns = ["المنطقة", "رقم الحساب", "اسم الزبون", "اسم المنتج", "وصف التعبئة"]
for col in categorical_columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le

# Debug: Check group sizes
group_counts = data.groupby(categorical_columns).size()
print("Group sizes:\n", group_counts)

# Define features and target (no lags)
features = categorical_columns + ["السنة", "الشهر", "month_sin", "month_cos"]
X = data[features]
y = data["الكمية باللتر"]

# Split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)

# Initialize and train LightGBM model
model = LGBMRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=6,
    random_state=42,
    categorical_feature=[features.index(col) for col in categorical_columns]
)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate model
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"Test RMSE: {rmse:.4f}")

# Visualize actual vs predicted
n_points = min(100, len(y_test))
plt.figure(figsize=(10, 6))
plt.plot(y_test.values[:n_points], label="Actual", marker="o")
plt.plot(y_pred[:n_points], label="Predicted", marker="x")
plt.legend()
plt.title("Actual vs Predicted الكمية باللتر")
plt.xlabel("Test Sample Index")
plt.ylabel("الكمية باللتر")
plt.show()

# Feature importance
feature_importance = model.feature_importances_
importance_df = pd.DataFrame({"Feature": features, "Importance": feature_importance})
importance_df = importance_df.sort_values("Importance", ascending=False)
print("Feature Importance:\n", importance_df)

# Save the model and encoders
joblib.dump(model, "lightgbm_model_no_lags.pkl")
for col, le in label_encoders.items():
    joblib.dump(le, f"label_encoder_{col}.pkl")
```

```python
# Prediction function
def predict_quantity(input_data, model_path="lightgbm_model_no_lags.pkl", encoder_paths=None):
    """
    Predict الكمية باللتر for new input data.
    
    Parameters:
    - input_data: DataFrame with columns [المنطقة, رقم الحساب, اسم الزبون, اسم المنتج, وصف التعبئة, السنة, الشهر]
    - model_path: Path to the trained LightGBM model
    - encoder_paths: Dict with paths to LabelEncoder files for categorical columns
    
    Returns:
    - Predicted الكمية باللتر
    """
    # Load model and encoders
    model = joblib.load(model_path)
    if encoder_paths is None:
        encoder_paths = {col: f"label_encoder_{col}.pkl" for col in categorical_columns}
    encoders = {col: joblib.load(path) for col, path in encoder_paths.items()}
    
    # Create a copy of input data
    input_df = input_data.copy()
    
    # Ensure data types
    input_df["المنطقة"] = input_df["المنطقة"].astype(str)
    input_df["رقم الحساب"] = input_df["رقم الحساب"].astype(str)
    input_df["اسم الزبون"] = input_df["اسم الزبون"].astype(str)
    input_df["اسم المنتج"] = input_df["اسم المنتج"].astype(str)
    input_df["وصف التعبئة"] = input_df["وصف التعبئة"].astype(str)
    input_df["السنة"] = input_df["السنة"].astype(int)
    input_df["الشهر"] = input_df["الشهر"].astype(int)
    
    # Encode categorical variables
    for col in categorical_columns:
        try:
            input_df[col] = encoders[col].transform(input_df[col])
        except ValueError as e:
            print(f"Error encoding {col}: {e}. Ensure input values exist in training data.")
            return None
    
    # Add seasonality features
    input_df["month_sin"] = np.sin(2 * np.pi * input_df["الشهر"] / 12)
    input_df["month_cos"] = np.cos(2 * np.pi * input_df["الشهر"] / 12)
    
    # Select features
    X_input = input_df[features]
    
    # Make predictions
    predictions = model.predict(X_input)
    return predictions

# Example usage of predict_quantity
example_input = pd.DataFrame({
    "المنطقة": ["2"],
    "رقم الحساب": ["170003"],
    "اسم الزبون": ["محطة كهرباءغرب طرابلس"],
    "اسم المنتج": ["وقود الديزل"],
    "وصف التعبئة": ["سائب/لتر"],
    "السنة": [2025],
    "الشهر": [1]
})
predictions = predict_quantity(example_input)
if predictions is not None:
    print("Predicted الكمية باللتر:", predictions)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from lightgbm import LGBMRegressor
import matplotlib.pyplot as plt
import joblib
import re
import unicodedata
from collections import Counter

# Function to clean feature names with duplicate handling
def clean_feature_name(name):
    """
    Replace Arabic characters and special symbols with ASCII-compatible strings.
    Handles both string and non-string (e.g., integer) column names.
    Preserves the target column 'الكمية باللتر'.
    """
    # Convert name to string
    name = str(name)
    
    # Skip cleaning for the target column
    if name == "الكمية باللتر":
        return name
    
    # Define specific mappings for Arabic terms
    replacements = {
        'المنطقة': 'Region',
        'رقم الحساب': 'AccountNumber', 
        'اسم الزبون': 'CustomerName',
        'اسم المنتج': 'ProductName',
        'وصف التعبئة': 'PackagingDescription',
        'السنة': 'Year',
        'الشهر': 'Month',
        'سائب': 'Bulk',
        'لتر': 'Liter',
        'وقود': 'Fuel',
        'الديزل': 'Diesel',
        'محطة': 'Station',
        'كهرباء': 'Electricity',
        'طرابلس': 'Tripoli',
        'غرب': 'West',
        'شرق': 'East',
        'شمال': 'North',
        'جنوب': 'South'
    }
    
    # Apply specific replacements first
    cleaned = name
    for arabic, ascii_val in replacements.items():
        cleaned = cleaned.replace(arabic, ascii_val)
    
    # Normalize Unicode characters to ASCII (for remaining Arabic text)
    try:
        cleaned = unicodedata.normalize('NFKD', cleaned).encode('ASCII', 'ignore').decode('ASCII')
    except:
        pass
    
    # Replace special characters
    cleaned = cleaned.replace('/', '_')
    cleaned = cleaned.replace(' ', '_')
    cleaned = cleaned.replace('-', '_')
    cleaned = cleaned.replace('.', '_')
    
    # Handle Unnamed columns
    if 'Unnamed' in cleaned:
        cleaned = cleaned.replace('Unnamed', 'ExtraCol')
    
    # Replace multiple underscores with a single one
    cleaned = re.sub(r'_+', '_', cleaned)
    
    # Remove non-alphanumeric characters except underscores
    cleaned = re.sub(r'[^A-Za-z0-9_]', '_', cleaned)
    
    # Remove leading/trailing underscores
    cleaned = cleaned.strip('_')
    
    # If empty after cleaning, create a meaningful fallback
    if not cleaned or cleaned == '':
        cleaned = f'Feature_{hash(name) % 10000}'
    
    return cleaned

def ensure_unique_feature_names(feature_names):
    """
    Ensure all feature names are unique by adding suffixes to duplicates.
    """
    name_counts = Counter(feature_names)
    unique_names = []
    counters = {}
    
    for name in feature_names:
        if name_counts[name] > 1:
            if name not in counters:
                counters[name] = 0
            else:
                counters[name] += 1
            unique_name = f"{name}_{counters[name]}" if counters[name] > 0 else name
            unique_names.append(unique_name)
        else:
            unique_names.append(name)
    
    return unique_names

# Load and preprocess data
data = pd.read_excel("ليبيا داتا.xlsx")  # Replace with your actual file path
print("Initial data shape:", data.shape)
print("Initial columns:", list(data.columns))

# Drop irrelevant columns (Unnamed and 2020 if empty or unnecessary)
irrelevant_cols = [col for col in data.columns if str(col).startswith('Unnamed') or str(col) == '2020']
data = data.drop(columns=irrelevant_cols, errors='ignore')

print("Shape after dropping irrelevant columns:", data.shape)
print("Remaining columns:", list(data.columns))

# Handle missing values
required_columns = ["المنطقة", "رقم الحساب", "اسم الزبون", "اسم المنتج", "وصف التعبئة", "السنة", "الشهر", "الكمية باللتر"]
existing_columns = [col for col in required_columns if col in data.columns]
data = data.dropna(subset=existing_columns)
print("Shape after dropping missing values:", data.shape)

# Ensure data types for existing columns
if "المنطقة" in data.columns:
    data["المنطقة"] = data["المنطقة"].astype(str)
if "رقم الحساب" in data.columns:
    data["رقم الحساب"] = data["رقم الحساب"].astype(str)
if "اسم الزبون" in data.columns:
    data["اسم الزبون"] = data["اسم الزبون"].astype(str)
if "اسم المنتج" in data.columns:
    data["اسم المنتج"] = data["اسم المنتج"].astype(str)
if "وصف التعبئة" in data.columns:
    data["وصف التعبئة"] = data["وصف التعبئة"].astype(str)
if "السنة" in data.columns:
    data["السنة"] = data["السنة"].astype(int)
if "الشهر" in data.columns:
    data["الشهر"] = data["الشهر"].astype(int)
if "الكمية باللتر" in data.columns:
    data["الكمية باللتر"] = data["الكمية باللتر"].clip(lower=0).astype(float)

# Add seasonality features if month column exists
if "الشهر" in data.columns:
    data["month_sin"] = np.sin(2 * np.pi * data["الشهر"] / 12)
    data["month_cos"] = np.cos(2 * np.pi * data["الشهر"] / 12)

# One-hot encode categorical variables
categorical_columns = [col for col in ["المنطقة", "رقم الحساب", "اسم الزبون", "اسم المنتج", "وصف التعبئة"] if col in data.columns]
if categorical_columns:
    data = pd.get_dummies(data, columns=categorical_columns, prefix=categorical_columns, dtype=int)
    print("Shape after one-hot encoding:", data.shape)

# Clean feature names and ensure uniqueness
original_names = list(data.columns)
cleaned_names = [clean_feature_name(col) for col in original_names]
unique_names = ensure_unique_feature_names(cleaned_names)

# Create mapping for debugging
name_mapping = dict(zip(original_names, unique_names))
print("Feature name mapping (first 10):")
for orig, clean in list(name_mapping.items())[:10]:
    print(f"  {orig} -> {clean}")

data.columns = unique_names
print("Final feature names (first 10):", list(data.columns)[:10])

# Verify target column exists
target_col = "الكمية باللتر"
if target_col not in data.columns:
    # Try to find it in the mapping
    possible_targets = [new_name for orig_name, new_name in name_mapping.items() if target_col in orig_name]
    if possible_targets:
        target_col = possible_targets[0]
        print(f"Using '{target_col}' as target column")
    else:
        raise KeyError(f"Target column '{target_col}' not found. Available columns: {list(data.columns)}")

# Define features and target
features = [col for col in data.columns if col != target_col and col != "time_idx"]
X = data[features]
y = data[target_col]

print(f"Number of features: {len(features)}")
print(f"Target column: {target_col}")
print(f"Data shape - X: {X.shape}, y: {y.shape}")

# Check for any remaining issues
print("Checking for duplicate feature names...")
duplicate_features = [name for name, count in Counter(features).items() if count > 1]
if duplicate_features:
    print(f"Warning: Found duplicate features: {duplicate_features}")
    # Remove duplicates by keeping first occurrence
    features = list(dict.fromkeys(features))  # Preserves order, removes duplicates
    X = X[features]
    print(f"After removing duplicates - Features: {len(features)}")

# Split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)

# Initialize and train LightGBM model
model = LGBMRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=6,
    num_leaves=31,
    random_state=42,
    verbosity=-1  # Reduce output
)

print("Training model...")
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate model
from sklearn.metrics import mean_squared_error, r2_score
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print(f"Test RMSE: {rmse:.4f}")
print(f"Test R²: {r2:.4f}")

# Visualize actual vs predicted
n_points = min(100, len(y_test))
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(y_test.values[:n_points], label="Actual", marker="o", alpha=0.7)
plt.plot(y_pred[:n_points], label="Predicted", marker="x", alpha=0.7)
plt.legend()
plt.title("Actual vs Predicted (Time Series)")
plt.xlabel("Test Sample Index")
plt.ylabel("الكمية باللتر")

plt.subplot(2, 2, 2)
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted (Scatter)")

plt.subplot(2, 2, 3)
residuals = y_test - y_pred
plt.scatter(y_pred, residuals, alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel("Predicted")
plt.ylabel("Residuals")
plt.title("Residual Plot")

plt.subplot(2, 2, 4)
plt.hist(residuals, bins=30, alpha=0.7)
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.title("Residual Distribution")

plt.tight_layout()
plt.show()

# Feature importance
feature_importance = model.feature_importances_
importance_df = pd.DataFrame({"Feature": features, "Importance": feature_importance})
importance_df = importance_df.sort_values("Importance", ascending=False)
print("Feature Importance (Top 15):")
print(importance_df.head(15))

# Plot feature importance
plt.figure(figsize=(10, 8))
top_features = importance_df.head(20)
plt.barh(range(len(top_features)), top_features["Importance"])
plt.yticks(range(len(top_features)), top_features["Feature"])
plt.xlabel("Importance")
plt.title("Top 20 Feature Importances")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# Save the model and feature names
joblib.dump(model, "lightgbm_model_cleaned.pkl")
joblib.dump(features, "feature_names_cleaned.pkl")
joblib.dump(name_mapping, "name_mapping.pkl")

print("Model and feature names saved successfully!")
print(f"Model performance - RMSE: {rmse:.4f}, R²: {r2:.4f}")

# Enhanced prediction function
def predict_quantity(input_data, model_path="lightgbm_model_cleaned.pkl", 
                    feature_names_path="feature_names_cleaned.pkl",
                    name_mapping_path="name_mapping.pkl"):
    """
    Predict الكمية باللتر for new input data, handling unseen categorical values.
    """
    # Load model, feature names, and name mapping
    model = joblib.load(model_path)
    feature_names = joblib.load(feature_names_path)
    name_mapping = joblib.load(name_mapping_path)
    
    # Create a copy of input data
    input_df = input_data.copy()
    
    # Ensure data types for existing columns
    for col in input_df.columns:
        if col in ["المنطقة", "رقم الحساب", "اسم الزبون", "اسم المنتج", "وصف التعبئة"]:
            input_df[col] = input_df[col].astype(str)
        elif col in ["السنة", "الشهر"]:
            input_df[col] = input_df[col].astype(int)
    
    # Add seasonality features if month exists
    if "الشهر" in input_df.columns:
        input_df["month_sin"] = np.sin(2 * np.pi * input_df["الشهر"] / 12)
        input_df["month_cos"] = np.cos(2 * np.pi * input_df["الشهر"] / 12)
    
    # One-hot encode categorical variables
    categorical_columns = [col for col in ["المنطقة", "رقم الحساب", "اسم الزبون", "اسم المنتج", "وصف التعبئة"] if col in input_df.columns]
    if categorical_columns:
        input_df = pd.get_dummies(input_df, columns=categorical_columns, prefix=categorical_columns, dtype=int)
    
    # Clean input feature names
    original_input_names = list(input_df.columns)
    cleaned_input_names = [clean_feature_name(col) for col in original_input_names]
    unique_input_names = ensure_unique_feature_names(cleaned_input_names)
    input_df.columns = unique_input_names
    
    # Align input features with training features
    missing_cols = [col for col in feature_names if col not in input_df.columns]
    extra_cols = [col for col in input_df.columns if col not in feature_names]
    
    # Add missing columns with zeros
    for col in missing_cols:
        input_df[col] = 0
    
    # Remove extra columns
    input_df = input_df.drop(columns=extra_cols, errors='ignore')
    
    # Ensure correct feature order
    X_input = input_df[feature_names]
    
    # Make predictions
    predictions = model.predict(X_input)
    return predictions

# Example usage
if len(data) > 0:
    # Create example input based on the first row of training data
    example_input = pd.DataFrame({
        col: [data[original_names[0]].iloc[0]] for col, original_col in 
        [(col, orig) for orig, col in name_mapping.items() if orig in categorical_columns[:5]][:3]
    })
    
    # Add numeric columns
    if "السنة" in data.columns:
        example_input["السنة"] = [2025]
    if "الشهر" in data.columns:
        example_input["الشهر"] = [1]
    
    try:
        predictions = predict_quantity(example_input)
        print(f"Predicted {target_col} for example input: {predictions}")
    except Exception as e:
        print(f"Prediction example failed: {e}")

print("Script completed successfully!")
```

```python
import pandas as pd
import numpy as np
import joblib
from collections import Counter
import re
import unicodedata

def clean_feature_name(name):
    """
    Replace Arabic characters and special symbols with ASCII-compatible strings.
    Handles both string and non-string (e.g., integer) column names.
    Preserves the target column 'الكمية باللتر'.
    """
    name = str(name)
    if name == "الكمية باللتر":
        return name
    replacements = {
        'المنطقة': 'Region',
        'رقم الحساب': 'AccountNumber', 
        'اسم الزبون': 'CustomerName',
        'اسم المنتج': 'ProductName',
        'وصف التعبئة': 'PackagingDescription',
        'السنة': 'Year',
        'الشهر': 'Month',
        'سائب': 'Bulk',
        'لتر': 'Liter',
        'وقود': 'Fuel',
        'الديزل': 'Diesel',
        'محطة': 'Station',
        'كهرباء': 'Electricity',
        'طرابلس': 'Tripoli',
        'غرب': 'West',
        'شرق': 'East',
        'شمال': 'North',
        'جنوب': 'South'
    }
    cleaned = name
    for arabic, ascii_val in replacements.items():
        cleaned = cleaned.replace(arabic, ascii_val)
    try:
        cleaned = unicodedata.normalize('NFKD', cleaned).encode('ASCII', 'ignore').decode('ASCII')
    except:
        pass
    cleaned = cleaned.replace('/', '_').replace(' ', '_').replace('-', '_').replace('.', '_')
    if 'Unnamed' in cleaned:
        cleaned = cleaned.replace('Unnamed', 'ExtraCol')
    cleaned = re.sub(r'_+', '_', cleaned)
    cleaned = re.sub(r'[^A-Za-z0-9_]', '_', cleaned)
    cleaned = cleaned.strip('_')
    if not cleaned or cleaned == '':
        cleaned = f'Feature_{hash(name) % 10000}'
    return cleaned

def ensure_unique_feature_names(feature_names):
    """
    Ensure all feature names are unique by adding suffixes to duplicates.
    """
    name_counts = Counter(feature_names)
    unique_names = []
    counters = {}
    for name in feature_names:
        if name_counts[name] > 1:
            if name not in counters:
                counters[name] = 0
            else:
                counters[name] += 1
            unique_name = f"{name}_{counters[name]}" if counters[name] > 0 else name
            unique_names.append(unique_name)
        else:
            unique_names.append(name)
    return unique_names

def predict_fuel_quantity(
    region=None,
    account_number=None,
    customer_name=None,
    product_name=None,
    packaging_description=None,
    year=None,
    month=None,
    model_path="lightgbm_model_cleaned.pkl",
    feature_names_path="feature_names_cleaned.pkl",
    name_mapping_path="name_mapping.pkl"
):
    """
    Predict fuel quantity (الكمية باللتر) based on user-provided input values.
    
    Parameters:
    -----------
    region : str, optional
        The region (المنطقة) of the fuel transaction.
    account_number : str, optional
        The account number (رقم الحساب) of the customer.
    customer_name : str, optional
        The customer name (اسم الزبون).
    product_name : str, optional
        The product name (اسم المنتج).
    packaging_description : str, optional
        The packaging description (وصف التعبئة).
    year : int, optional
        The year (السنة) of the transaction.
    month : int, optional
        The month (الشهر) of the transaction (1-12).
    model_path : str, optional
        Path to the saved LightGBM model file.
    feature_names_path : str, optional
        Path to the saved feature names file.
    name_mapping_path : str, optional
        Path to the saved feature name mapping file.
    
    Returns:
    --------
    float
        Predicted fuel quantity (الكمية باللتر).
    
    Example:
    --------
    prediction = predict_fuel_quantity(
        region="طرابلس",
        account_number="12345",
        customer_name="شركة الزبون",
        product_name="الديزل",
        packaging_description="سائب",
        year=2025,
        month=1
    )
    print(f"Predicted Quantity: {prediction}")
    """
    # Load model, feature names, and name mapping
    try:
        model = joblib.load(model_path)
        feature_names = joblib.load(feature_names_path)
        name_mapping = joblib.load(name_mapping_path)
    except FileNotFoundError as e:
        raise FileNotFoundError(f"Model or feature files not found: {e}")

    # Create input dictionary with provided values
    input_data = {}
    if region is not None:
        input_data["المنطقة"] = str(region)
    if account_number is not None:
        input_data["رقم الحساب"] = str(account_number)
    if customer_name is not None:
        input_data["اسم الزبون"] = str(customer_name)
    if product_name is not None:
        input_data["اسم المنتج"] = str(product_name)
    if packaging_description is not None:
        input_data["وصف التعبئة"] = str(packaging_description)
    if year is not None:
        try:
            input_data["السنة"] = int(year)
        except ValueError:
            raise ValueError("Year must be an integer.")
    if month is not None:
        try:
            month = int(month)
            if month < 1 or month > 12:
                raise ValueError("Month must be between 1 and 12.")
            input_data["الشهر"] = month
        except ValueError:
            raise ValueError("Month must be an integer between 1 and 12.")

    if not input_data:
        raise ValueError("At least one input value must be provided.")

    # Convert to DataFrame
    input_df = pd.DataFrame([input_data])

    # Add seasonality features if month is provided
    if "الشهر" in input_df.columns:
        input_df["month_sin"] = np.sin(2 * np.pi * input_df["الشهر"] / 12)
        input_df["month_cos"] = np.cos(2 * np.pi * input_df["الشهر"] / 12)

    # One-hot encode categorical variables
    categorical_columns = [col for col in ["المنطقة", "رقم الحساب", "اسم الزبون", "اسم المنتج", "وصف التعبئة"] if col in input_df.columns]
    if categorical_columns:
        input_df = pd.get_dummies(input_df, columns=categorical_columns, prefix=categorical_columns, dtype=int)

    # Clean input feature names
    original_input_names = list(input_df.columns)
    cleaned_input_names = [clean_feature_name(col) for col in original_input_names]
    unique_input_names = ensure_unique_feature_names(cleaned_input_names)
    input_df.columns = unique_input_names

    # Align input features with training features
    missing_cols = [col for col in feature_names if col not in input_df.columns]
    extra_cols = [col for col in input_df.columns if col not in feature_names]
    for col in missing_cols:
        input_df[col] = 0
    input_df = input_df.drop(columns=extra_cols, errors='ignore')

    # Ensure correct feature order
    X_input = input_df[feature_names]

    # Make prediction
    prediction = model.predict(X_input)[0]
    return max(0, prediction)  # Ensure non-negative prediction

# Example usage
if __name__ == "__main__":
    try:
        prediction = predict_fuel_quantity(
            region="1",
            account_number="170127",
            customer_name=" محطةكهرباء مصراتةالغازية",
            product_name="وقود الديزل",
            packaging_description="سائب/لتر",
            year=2020,
            month=1
        )
        print(f"Predicted الكمية باللتر: {prediction:.2f} liters")
    except Exception as e:
        print(f"Error: {e}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from xgboost import XGBRegressor
import matplotlib.pyplot as plt
import joblib
import re
import unicodedata
from collections import Counter
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, make_scorer

# --- الدوال المساعدة ---
def clean_feature_name(name):
    """
    Replace Arabic characters and special symbols with ASCII-compatible strings.
    Handles both string and non-string (e.g., integer) column names.
    Preserves the target column 'الكمية باللتر'.
    """
    name = str(name)
    if name == "الكمية باللتر":
        return name
    replacements = {
        'المنطقة': 'Region', 'رقم الحساب': 'AccountNumber', 'اسم الزبون': 'CustomerName',
        'اسم المنتج': 'ProductName', 'وصف التعبئة': 'PackagingDescription', 'السنة': 'Year',
        'الشهر': 'Month', 'سائب': 'Bulk', 'لتر': 'Liter', 'وقود': 'Fuel', 'الديزل': 'Diesel',
        'محطة': 'Station', 'كهرباء': 'Electricity', 'طرابلس': 'Tripoli', 'غرب': 'West',
        'شرق': 'East', 'شمال': 'North', 'جنوب': 'South'
    }
    cleaned = name
    for arabic, ascii_val in replacements.items():
        cleaned = cleaned.replace(arabic, ascii_val)
    try:
        cleaned = unicodedata.normalize('NFKD', cleaned).encode('ASCII', 'ignore').decode('ASCII')
    except:
        pass
    cleaned = cleaned.replace('/', '_').replace(' ', '_').replace('-', '_').replace('.', '_')
    if 'Unnamed' in cleaned:
        cleaned = cleaned.replace('Unnamed', 'ExtraCol')
    cleaned = re.sub(r'_+', '_', cleaned)
    cleaned = re.sub(r'[^A-Za-z0-9_]', '_', cleaned)
    cleaned = cleaned.strip('_')
    if not cleaned or cleaned == '':
        cleaned = f'Feature_{hash(name) % 10000}'
    return cleaned

def ensure_unique_feature_names(feature_names):
    """
    Ensure all feature names are unique by adding suffixes to duplicates.
    """
    name_counts = Counter(feature_names)
    unique_names = []
    counters = {}
    for name in feature_names:
        if name_counts[name] > 1:
            if name not in counters:
                counters[name] = 0
            else:
                counters[name] += 1
            unique_name = f"{name}_{counters[name]}" if counters[name] > 0 else name
            unique_names.append(unique_name)
        else:
            unique_names.append(name)
    return unique_names

def custom_accuracy_score(y_true, y_pred, tolerance_percentage=0.10):
    """
    Calculates the percentage of predictions that are within a specified tolerance
    of the actual value.
    """
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    
    # Ensure y_true is at least 1 to avoid issues with percentage calculation for very small actuals
    y_true_safe = np.maximum(1.0, y_true) 
    
    absolute_diff = np.abs(y_true - y_pred)
    tolerance_threshold = tolerance_percentage * y_true_safe
    
    correct_predictions = np.sum(absolute_diff <= tolerance_threshold)
    accuracy = correct_predictions / len(y_true) * 100
    return accuracy

# --- تحميل ومعالجة البيانات ---
data = pd.read_excel("ليبيا داتا.xlsx")
print("Initial data shape:", data.shape)
print("Initial columns:", list(data.columns))

irrelevant_cols = [col for col in data.columns if str(col).startswith('Unnamed') or str(col) == '2020']
data = data.drop(columns=irrelevant_cols, errors='ignore')
print("Shape after dropping irrelevant columns:", data.shape)
print("Remaining columns:", list(data.columns))

required_columns = ["المنطقة", "رقم الحساب", "اسم الزبون", "اسم المنتج", "وصف التعبئة", "السنة", "الشهر", "الكمية باللتر"]
existing_columns = [col for col in required_columns if col in data.columns]
data = data.dropna(subset=existing_columns)
print("Shape after dropping missing values:", data.shape)

if "المنطقة" in data.columns: data["المنطقة"] = data["المنطقة"].astype(str)
if "رقم الحساب" in data.columns: data["رقم الحساب"] = data["رقم الحساب"].astype(str)
if "اسم الزبون" in data.columns: data["اسم الزبون"] = data["اسم الزبون"].astype(str)
if "اسم المنتج" in data.columns: data["اسم المنتج"] = data["اسم المنتج"].astype(str)
if "وصف التعبئة" in data.columns: data["وصف التعبئة"] = data["وصف التعبئة"].astype(str)
if "السنة" in data.columns: data["السنة"] = data["السنة"].astype(int)
if "الشهر" in data.columns: data["الشهر"] = data["الشهر"].astype(int)

# --- معالجة القيم السلبية في الكمية باللتر قبل التحويل اللوغاريتمي ---
# التأكد من أن جميع قيم 'الكمية باللتر' موجبة تمامًا لأن log(0) غير معرف
if "الكمية باللتر" in data.columns:
    data["الكمية باللتر"] = data["الكمية باللتر"].clip(lower=1).astype(float) # استخدم 1 بدلاً من 0 لتجنب log(0)
    # --- التحويل اللوغاريتمي للعمود المستهدف ---
    # استخدام log1p (log(1+x)) وهو أفضل للتعامل مع القيم الصغيرة جداً والقيم الصفرية المحتملة لو كانت موجودة
    data["الكمية باللتر_log"] = np.log1p(data["الكمية باللتر"])
    print("Applied log1p transformation to 'الكمية باللتر'.")


if "الشهر" in data.columns:
    data["month_sin"] = np.sin(2 * np.pi * data["الشهر"] / 12)
    data["month_cos"] = np.cos(2 * np.pi * data["الشهر"] / 12)

categorical_columns = [col for col in ["المنطقة", "رقم الحساب", "اسم الزبون", "اسم المنتج", "وصف التعبئة"] if col in data.columns]
if categorical_columns:
    data = pd.get_dummies(data, columns=categorical_columns, prefix=categorical_columns, dtype=int)
    print("Shape after one-hot encoding:", data.shape)

original_names = list(data.columns)
cleaned_names = [clean_feature_name(col) for col in original_names]
unique_names = ensure_unique_feature_names(cleaned_names)
name_mapping = dict(zip(original_names, unique_names))
print("Feature name mapping (first 10):")
for orig, clean in list(name_mapping.items())[:10]:
    print(f"  {orig} -> {clean}")

data.columns = unique_names
print("Final feature names (first 10):", list(data.columns)[:10])

# --- تعديل اسم العمود المستهدف ليصبح الاسم بعد تنظيف الأسماء ---
target_col_original = "الكمية باللتر"
# المشكلة كانت هنا، حيث أن 'الكمية باللتر_log' تحول إلى 'Liter_log' بعد تنظيف الأسماء
target_col = "Liter_log" # الاسم الصحيح بعد عملية التنظيف

if target_col not in data.columns:
    raise KeyError(f"Target column '{target_col}' not found. Ensure log transformation was applied and name is correct after cleaning.")

features = [col for col in data.columns if col not in [target_col, target_col_original, "time_idx", "الكمية باللتر"]] # استبعاد العمود الأصلي
X = data[features]
y = data[target_col] # Y هنا هي القيم اللوغاريتمية

print(f"Number of features: {len(features)}")
print(f"Target column for training: {target_col}")
print(f"Data shape - X: {X.shape}, y: {y.shape}")

print("Checking for duplicate feature names...")
duplicate_features = [name for name, count in Counter(features).items() if count > 1]
if duplicate_features:
    print(f"Warning: Found duplicate features: {duplicate_features}")
    features = list(dict.fromkeys(features))
    X = X[features]
    print(f"After removing duplicates - Features: {len(features)}")

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)

# --- ضبط المعاملات باستخدام GridSearchCV ---
print("\n--- بدأ عملية ضبط معاملات XGBoost (Grid Search Cross-Validation) ---")

# تم تصحيح الخطأ الإملائي هنا: 'reg:squarederror' بدلاً من 'reg:squareerror'
xgb = XGBRegressor(objective='reg:squarederror', random_state=42, n_jobs=-1)

# نطاق أوسع للمعاملات
param_grid = {
    'n_estimators': [400, 600, 800], # عدد الأشجار
    'learning_rate': [0.01, 0.03, 0.05], # معدل التعلم
    'max_depth': [7, 9, 11], # أقصى عمق للشجرة
    'subsample': [0.7, 0.8], # نسبة العينات لكل شجرة
    'colsample_bytree': [0.7, 0.8] # نسبة الميزات لكل شجرة
}

# هنا بنستخدم MAPE أو MAE كـ scoring للموديل في الـ Grid Search
# لأن مقياس الدقة المخصص (custom_accuracy_score) بيتعامل مع القيم الأصلية، مش اللوغاريتمية.
# وبالتالي بنركز على مقاييس الخطأ القياسية هنا.
grid_search = GridSearchCV(estimator=xgb, param_grid=param_grid, 
                           scoring='neg_mean_absolute_error', # عشان يختار أفضل موديل بأقل MAE
                           cv=3, verbose=2, n_jobs=-1)

grid_search.fit(X_train, y_train)

print("\n--- انتهت عملية ضبط المعاملات ---")
print(f"أفضل معاملات تم العثور عليها: {grid_search.best_params_}")
print(f"أفضل MAE (باستخدام القيمة السالبة) على بيانات التحقق المتقاطع: {grid_search.best_score_:.4f}")

# استخدام أفضل موديل تم العثور عليه
model = grid_search.best_estimator_

print("\nTraining XGBoost model with best hyperparameters...")

# Make predictions
y_pred_log = model.predict(X_test)

# --- عكس التحويل اللوغاريتمي للنتائج ---
# np.expm1 هو عكس np.log1p (e^x - 1)
y_pred = np.expm1(y_pred_log)
y_test_original = np.expm1(y_test) # نحصل على القيم الأصلية للاختبار للمقارنة

# --- تقييم الموديل بالمقاييس الأصلية ---
rmse = np.sqrt(mean_squared_error(y_test_original, y_pred))
r2 = r2_score(y_test_original, y_pred)
mae = mean_absolute_error(y_test_original, y_pred)

# حساب MAPE (تجنب القسمة على صفر)
# نستخدم يدويا للتعامل مع القيم الصفرية لتجنب NaNs أو Infs
mape = np.mean(np.abs((y_test_original - y_pred) / np.maximum(1.0, y_test_original))) * 100 
if np.isinf(mape) or np.isnan(mape):
    mape_display = "N/A (contains zero actuals)"
else:
    mape_display = f"{mape:.2f}%"

# حساب الدقة المخصصة
accuracy_percentage = custom_accuracy_score(y_test_original, y_pred)

print(f"\n--- نتائج تقييم موديل XGBoost المحسّن مع التحويل اللوغاريتمي ---")
print(f"Test RMSE: {rmse:.4f}")
print(f"Test R²: {r2:.4f}")
print(f"Test MAE: {mae:.4f} liters")
print(f"Test MAPE: {mape_display}")
print(f"Test Accuracy (within 10% tolerance): {accuracy_percentage:.2f}%")
print(f"----------------------------------------\n")


# --- تصور النتائج (باستخدام القيم الأصلية) ---
n_points = min(100, len(y_test_original))
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(y_test_original.values[:n_points], label="Actual", marker="o", alpha=0.7)
plt.plot(y_pred[:n_points], label="Predicted", marker="x", alpha=0.7)
plt.legend()
plt.title("Actual vs Predicted (Time Series)")
plt.xlabel("Test Sample Index")
plt.ylabel("الكمية باللتر")

plt.subplot(2, 2, 2)
plt.scatter(y_test_original, y_pred, alpha=0.5)
plt.plot([y_test_original.min(), y_test_original.max()], [y_test_original.min(), y_test_original.max()], 'r--', lw=2)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted (Scatter)")

plt.subplot(2, 2, 3)
residuals = y_test_original - y_pred
plt.scatter(y_pred, residuals, alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel("Predicted")
plt.ylabel("Residuals")
plt.title("Residual Plot")

plt.subplot(2, 2, 4)
plt.hist(residuals, bins=30, alpha=0.7)
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.title("Residual Distribution")

plt.tight_layout()
plt.show()

# --- أهمية الميزات ---
feature_importance = model.feature_importances_
importance_df = pd.DataFrame({"Feature": features, "Importance": feature_importance})
importance_df = importance_df.sort_values("Importance", ascending=False)
print("Feature Importance (Top 15):")
print(importance_df.head(15))

plt.figure(figsize=(10, 8))
top_features = importance_df.head(20)
plt.barh(range(len(top_features)), top_features["Importance"])
plt.yticks(range(len(top_features)), top_features["Feature"])
plt.xlabel("Importance")
plt.title("Top 20 Feature Importances")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# --- حفظ الموديل وأسماء الميزات ---
joblib.dump(model, "xgboost_model_final_tuned.pkl") # اسم جديد ليدل على التحسين النهائي
joblib.dump(features, "feature_names_cleaned.pkl")
joblib.dump(name_mapping, "name_mapping.pkl")

print("Model and feature names saved successfully!")
print(f"Final Model performance - RMSE: {rmse:.4f}, R²: {r2:.4f}, MAE: {mae:.4f}, MAPE: {mape_display}, Accuracy: {accuracy_percentage:.2f}%")


# --- دالة التوقع النهائية ---
def predict_fuel_quantity(
    region=None, account_number=None, customer_name=None, product_name=None,
    packaging_description=None, year=None, month=None,
    model_path="xgboost_model_final_tuned.pkl", # تحديث اسم الموديل هنا
    feature_names_path="feature_names_cleaned.pkl",
    name_mapping_path="name_mapping.pkl"
):
    """
    Predict fuel quantity (الكمية باللتر) based on user-provided input values
    using the logarithmically transformed model.
    """
    try:
        model = joblib.load(model_path)
        feature_names = joblib.load(feature_names_path)
        name_mapping = joblib.load(name_mapping_path)
    except FileNotFoundError as e:
        raise FileNotFoundError(f"Model or feature files not found: {e}")

    input_data = {}
    if region is not None: input_data["المنطقة"] = str(region)
    if account_number is not None: input_data["رقم الحساب"] = str(account_number)
    if customer_name is not None: input_data["اسم الزبون"] = str(customer_name)
    if product_name is not None: input_data["اسم المنتج"] = str(product_name)
    if packaging_description is not None: input_data["وصف التعبئة"] = str(packaging_description)
    if year is not None:
        try: input_data["السنة"] = int(year)
        except ValueError: raise ValueError("Year must be an integer.")
    if month is not None:
        try:
            month = int(month)
            if month < 1 or month > 12: raise ValueError("Month must be between 1 and 12.")
            input_data["الشهر"] = month
        except ValueError: raise ValueError("Month must be an integer between 1 and 12.")

    if not input_data:
        raise ValueError("At least one input value must be provided.")

    input_df = pd.DataFrame([input_data])

    if "الشهر" in input_df.columns:
        input_df["month_sin"] = np.sin(2 * np.pi * input_df["الشهر"] / 12)
        input_df["month_cos"] = np.cos(2 * np.pi * input_df["الشهر"] / 12)

    categorical_columns = [col for col in ["المنطقة", "رقم الحساب", "اسم الزبون", "اسم المنتج", "وصف التعبئة"] if col in input_df.columns]
    if categorical_columns:
        input_df = pd.get_dummies(input_df, columns=categorical_columns, prefix=categorical_columns, dtype=int)

    original_input_names = list(input_df.columns)
    cleaned_input_names = [clean_feature_name(col) for col in original_input_names]
    unique_input_names = ensure_unique_feature_names(cleaned_input_names)
    input_df.columns = unique_input_names

    missing_cols = [col for col in feature_names if col not in input_df.columns]
    extra_cols = [col for col in input_df.columns if col not in feature_names]
    for col in missing_cols:
        input_df[col] = 0
    input_df = input_df.drop(columns=extra_cols, errors='ignore')

    X_input = input_df[feature_names]
    
    # Make prediction on log-transformed target
    prediction_log = model.predict(X_input)[0]
    
    # Reverse the log transformation to get the actual quantity
    prediction_original = np.expm1(prediction_log) # expm1 is inverse of log1p
    
    return max(0, prediction_original) # Ensure non-negative prediction

# --- مثال على الاستخدام ---
if len(data) > 0:
    example_input = pd.DataFrame({
        "المنطقة": ["1"],
        "رقم الحساب": ["170003"],
        "اسم الزبون": ["محطة كهرباءغرب طرابلس"],
        "اسم المنتج": ["وقود الديزل"],
        "وصف التعبئة": ["سائب/لتر"],
        "السنة": [2020],
        "الشهر": [1]
    })
    
    try:
        predictions = predict_fuel_quantity(example_input)
        print(f"Predicted الكمية باللتر for example input: {predictions:.2f} liters")
    except Exception as e:
        print(f"Prediction example failed: {e}")

print("Script completed successfully!")
```
